```python
#pip install geopy    (help to get latitude and longitude)
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='app',timeout=None)
geolocator
```

```python
lat = []
lon = []

for name in df1['Panchyat']:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
lat
```

```python
df1['lat'] = lat
df1['lon'] = lon
```
